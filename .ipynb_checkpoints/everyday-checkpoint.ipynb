```python
#讀資料庫並分群
def train_mode(engine):
    sql="select * from table2 "
    df=pd.read_sql(sql,engine)
    personality= df['personality_score'].values.tolist()
    interest= df['interest_score'].values.tolist()
    tfidf = np.column_stack((personality,interest))  #toarray
    ##分三群
    k = 3
    Kmeans_cluster = KMeans()
    content_cluster_result = Kmeans_cluster.kmeans_plus_plus(tfidf, k)

    cluster = {0: [], 1: [], 2: []}
    index = 0
    for result in content_cluster_result:
        if int(result) == 0:
            group = 0
            cluster[group].append(tfidf[index].tolist())
        elif int(result) == 1:
            group = 1
            cluster[group].append(tfidf[index].tolist())
        elif int(result) == 2:
            group = 2
            cluster[group].append(tfidf[index].tolist())
        index += 1
    return cluster
def center_re(cluster):
    center_records = {}
    for n in range(3):
        center_records[n] = (sum(np.array(cluster[n])) / len(cluster[n])).tolist()
    return center_records
#算中心點
def center(engine,cluster): 
    sql="select * from table2 " 
    df = pd.read_sql(sql,engine)
    center_records = center_re(cluster)   
    new_data = []
    for i in range(len(cluster)):
        for j in range(len(cluster[i])):
            for k in range(len(df)):
                if df.loc[k]['personality_score'] == cluster[i][j][0] and df.loc[k]['interest_score'] == cluster[i][j][1]:
                    a={}
                    a['name'] = df['nickname'][k]
                    a['email'] = df['email'][k]
                    a['gender'] = df['gender'][k]
                    a['r_gender'] = df['r_gender'][k]
                    a['age'] = df['age'][k]
                    a['r_age'] = df['r_age'][k]
                    a['weight'] = df['weight'][k] 
                    a['r_weight'] = df['r_weight'][k] 
                    a['height'] = df['height'][k] 
                    a['r_height'] = df['r_height'][k]
                    a['personality'] = df['personality'][k]
                    a['zodiac'] = df['zodiac'][k]
                    a['job'] = df['job'][k]
                    a['test1'] = df['test1'][k]
                    a['test2'] = df['test2'][k]
                    a['test3'] = df['test3'][k]
                    a['test4_1'] = df['test4_1'][k]
                    a['test4_2'] = df['test4_2'][k]
                    a['test4_3'] = df['test4_3'][k]
                    a['test4_4'] = df['test4_4'][k]
                    a['r_personality'] = df['r_personality'][k]                 
                    a['personality_score'] = df['personality_score'][k]
                    a['interest_score'] = df['interest_score'][k]
                    a['mure'] = i
                    a['center'] = center_records[i]
                    new_data.append(a)
    new_df=pd.DataFrame(new_data)
    new_df.to_csv("new_data.csv",index=True, index_label="id")
    df = pd.read_csv("new_data.csv", sep=',')
    df.to_sql('table2', engine, index= False,if_exists = "replace")
```
